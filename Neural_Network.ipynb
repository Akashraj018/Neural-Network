1. What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

Deep Learning is a subfield of machine learning that uses artificial neural networks with many layers (deep neural networks) to automatically learn complex patterns from large amounts of data. It is especially effective for tasks such as image recognition, speech processing, natural language understanding, and game playing.

Evolution of Deep Learning
1. Early Neural Networks (1940s–1980s)
 Concepts like the perceptron and basic neural networks were introduced.Progress was limited by low computational power and lack of large datasets.
2. Revival with Backpropagation (1980s–1990s)
The backpropagation algorithm enabled training of multi-layer networks.Still limited in practice due to hardware and data constraints.
3. Modern Deep Learning Boom (2000s–Present)
Availability of big data, GPUs, and improved algorithms (e.g., ReLU, dropout, batch normalization).
Breakthroughs in computer vision, speech recognition, and NLP (e.g., CNNs, RNNs, Transformers).

2.  Explain the basic architecture and functioning of a Perceptron. What are its limitations?

A Perceptron is the simplest form of an artificial neural network and is used for binary classification. It consists of the following components:
 - Input Layer- Takes multiple input features x1,x2,x3,x4,.......xn
 - Weights - Each input is associated with a weight w1,w2,w3,....wn that represents its importance.
 - Bias - A bias term b is added to shift the decision boundary.
 - Summation Unit - Computes the weighted sum:
              z=i=1∑wixi + b
  - Activation Function
  Typically a step (threshold) function that produces the output:
   Y = {1,0  if z >- 0 otherwise

Functioning of a Perceptron
Input values are multiplied by their corresponding weights.
The weighted inputs are summed along with the bias.
The result is passed through the activation function.
The output represents the predicted class (0 or 1).

Limitations of a Perceptron
- Only Linearly Separable Problems - Cannot solve non-linear problems (e.g., XOR problem).
- Single-Layer Model - Lacks hidden layers, limiting its representational power.
- Binary Output Only - Standard perceptron handles only binary classification.
- Simple Activation Function - Step function is not differentiable, limiting advanced optimization methods.



3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

An activation function determines whether a neuron should be activated by introducing non-linearity into the network. Without activation functions, a neural network would behave like a linear model regardless of the number of layers. Activation functions enable neural networks to:

Learn complex, non-linear relationships
Decide which information to pass forward
Improve model expressiveness and performance

Key Differences Explained
Sigmoid - Good for binary classification outputs.
Suffers from vanishing gradient problem, slowing learning in deep networks.
Tanh -Zero-centered outputs help with faster convergence than sigmoid.
Still affected by vanishing gradients for large input values.
ReLU (Rectified Linear Unit) -
Simple and computationally efficient.
Reduces vanishing gradient issues and enables deeper networks.
Suffers from the dying ReLU problem (neurons output zero for all inputs).

4.  What is the difference between Loss function and Cost function in neural networks? Provide examples.

In neural networks, loss function and cost function are closely related terms, but they differ in scope.

Loss Function-
A loss function measures the error for a single training example.
It quantifies how far the predicted output is from the true output for one data point.
Used during training to guide weight updates.

Cost Function-
A cost function is the average (or sum) of the loss function over the entire training dataset.
It represents the overall performance of the neural network.
Optimization algorithms aim to minimize the cost function.



5.  What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.  

An optimizer is an algorithm used to update the weights and biases of a neural network during training in order to minimize the loss (cost) function. Optimizers determine how and how fast a neural network learns by controlling the direction and size of parameter updates.

Key roles of optimizers:
Minimize the loss function
Speed up convergence
Avoid local minima or saddle points
Handle noisy and sparse gradients efficiently

Explanation of Each Optimizer
1. Gradient Descent
Updates weights in the direction of the negative gradient of the loss.
Limitations: Slow convergence, sensitive to learning rate, inefficient for large datasets.
2. RMSprop (Root Mean Square Propagation)
Maintains a moving average of squared gradients.
Advantages: Handles non-stationary objectives, works well for RNNs.
Limitation: Does not include bias correction.
3. Adam (Adaptive Moment Estimation)
Combines Momentum (first moment) and RMSprop (second moment).
Advantages: Fast convergence, robust, widely used.
Limitation: May sometimes converge to suboptimal solutions.

Optimizers are crucial for efficient training of neural networks.
Gradient Descent is simple but slow.
RMSprop adapts learning rates to improve stability.
Adam combines the strengths of momentum and adaptive learning rates, making it the most popular choice in modern deep learning.

```python
import numpy as np

# Input data for AND gate
# X: inputs, y: expected outputs
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1
epochs = 10

# Activation function (step function)
def step_function(z):
    return 1 if z >= 0 else 0

# Training the perceptron
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        # Linear combination
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = step_function(z)

        # Weight and bias update
        error = y[i] - y_pred
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Prediction: {y_pred}, Weights: {weights}, Bias: {bias}")
    print()

# Testing the trained perceptron
print("Final Results:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    output = step_function(z)
    print(f"Input: {X[i]} -> Output: {output}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define input range
x = np.linspace(-10, 10, 400)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot the activation functions
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label="Sigmoid", linewidth=2)
plt.plot(x, y_relu, label="ReLU", linewidth=2)
plt.plot(x, y_tanh, label="Tanh", linewidth=2)

plt.title("Activation Functions Visualization")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten images (28x28 -> 784)
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=128,
    verbose=1
)

# Print training accuracy
training_accuracy = history.history['accuracy'][-1]
print(f"Training Accuracy: {training_accuracy * 100:.2f}%")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten images
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=128,
    validation_data=(X_test, y_test),
    verbose=1
)

# Plot Loss curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot Accuracy curves
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.show()
```
